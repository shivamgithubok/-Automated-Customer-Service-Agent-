```python
# Updated script for your environment

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Function to set up and scrape the webpage
def scrape_react_page(url):
    # Set up Chrome options for headless mode (no GUI)
    chrome_options = Options()
    chrome_options.add_argument('--headless')  # Run Chrome in headless mode (no GUI)
    chrome_options.add_argument('--no-sandbox')  # Disable sandboxing for Chromium
    chrome_options.add_argument('--disable-dev-shm-usage')  # Handle resource issues

    # Set up the ChromeDriver Service
    service = Service(ChromeDriverManager().install())

    # Initialize the driver with the Service and options
    driver = webdriver.Chrome(service=service, options=chrome_options)

    try:
        # Open the webpage URL
        driver.get(url)

        # Wait for the page to load (you can replace time.sleep with explicit waits for better control)
        driver.implicitly_wait(10)  # Wait for up to 10 seconds for page content to load

        # Get the fully rendered HTML page source
        page_html = driver.page_source

        # Print the rendered HTML
        print(page_html)

    finally:
        # Close the browser after scraping
        driver.quit()

# URL of the React page you want to scrape
url = "https://www.linkedin.com/in/shivam1-chaudhary/"

# Call the function to scrape the page
scrape_react_page(url)
```
